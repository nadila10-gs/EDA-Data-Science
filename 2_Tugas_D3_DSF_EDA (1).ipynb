````python
# Mengatasi missing value
for column in df.columns:
    if df[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # Jika kolom bertipe numerik, isi dengan mean
        df[column].fillna(df[column].mean(), inplace=True)
````

# EDA

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
````

## Membuat kolom baru yaitu ROI (Return on Investment)

````python
df['BoxOffice_USD'] = (
    df['BoxOffice_USD']
    .astype(str)
    .str.strip()
    .replace('', np.nan)
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

df['ROI'] = df['BoxOffice_USD'] / df['Budget_USD']
````

````python
df['ROI'] = df['BoxOffice_USD'] / df['Budget_USD']
df_valid = df.dropna(subset=['BoxOffice_USD', 'Budget_USD'])
````

````python
numeric_df = df.select_dtypes(include=['number']).copy()
# MATRIKS KORELASI
corr_matrix = numeric_df.corr(method='pearson')

# PLOT HEATâ€‘MAP
plt.figure(figsize=(12, 10))
sns.set(font_scale=0.9)
heatmap = sns.heatmap(
    corr_matrix,
    cmap='coolwarm',                  # skema warna biruâ€‘merah
    annot=True,                       # tampilkan angka korelasi di tiap sel
    fmt='.2f',                        # dua angka di belakang koma
    vmin=-1, vmax=1,                  # rentang warna penuh â€“1 s.d. +1
    linewidths=.5,                    # garis pemisah antar sel
    cbar_kws={'label': 'Koefisien Korelasi'}
)

plt.title('Heatâ€‘Map Korelasi Fitur')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()                     # atur padding otomatis
plt.show()
````

````python
pair_df = df[['Budget_USD', 'ROI']].dropna()

# HITUNG KORELASI + pâ€‘VALUE
r, p = pearsonr(pair_df['Budget_USD'], pair_df['ROI'])
print(f'Korelasi Budget â†” ROI : r = {r:.2f}   |   pâ€‘value = {p:.4g}')

# SCATTERâ€‘PLOT + REGRESI
sns.set(style='whitegrid', font_scale=1.1)
plt.figure(figsize=(7, 6))

sns.regplot(
    data=pair_df,
    x='Budget_USD',
    y='ROI',
    scatter_kws={'alpha': 0.65},
    line_kws={'linewidth': 2}
)

plt.title('Budget vs ROI')
plt.xlabel('Budget (USD)')
plt.ylabel('ROI')
plt.tight_layout()
plt.show()
````

````python
pair_df = df[['BoxOffice_USD', 'ROI']].dropna()

# HITUNG KORELASI + pâ€‘VALUE
r, p = pearsonr(pair_df['BoxOffice_USD'], pair_df['ROI'])
print(f'Korelasi BoxOffice_USD â†” ROI : r = {r:.2f}   |   pâ€‘value = {p:.4g}')

# SCATTERâ€‘PLOT + REGRESI
sns.set(style='whitegrid', font_scale=1.1)
plt.figure(figsize=(7, 6))

sns.regplot(
    data=pair_df,
    x='BoxOffice_USD',
    y='ROI',
    scatter_kws={'alpha': 0.65},
    line_kws={'linewidth': 2}
)

plt.title('BoxOffice_USD vs ROI')
plt.xlabel('BoxOffice_USD')
plt.ylabel('ROI')
plt.tight_layout()
plt.show()
````

````python
heat_data = pd.crosstab(df['Genre'], df['Content_Rating'])

plt.figure(figsize=(12, 6))
sns.heatmap(heat_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Genre vs Content Rating")
plt.tight_layout()
plt.show()
````

````python
rating_vote_stats = (df.groupby('Rating_bin')['Votes']
                       .agg(['mean', 'count', 'std'])
                       .assign(se=lambda d: d['std'] / np.sqrt(d['count']),
                               ci_low=lambda d: d['mean'] - 1.96*d['se'],
                               ci_hi=lambda d: d['mean'] + 1.96*d['se'])
                       .reset_index())
plt.figure(figsize=(11, 4))
x = np.arange(len(rating_vote_stats))

# Line chart
plt.plot(x, rating_vote_stats['mean'], marker='o', color='darkblue', label='Mean Votes')

# CI band
plt.fill_between(x, rating_vote_stats['ci_low'], rating_vote_stats['ci_hi'],
                 color='blue', alpha=0.2, label='95% CI')

# Label sumbu X
plt.xticks(x, rating_vote_stats['Rating_bin'].astype(str), rotation=45)

plt.yscale('log')
plt.xlabel('Rating (binned)')
plt.ylabel('Mean Votes (log scale)')
plt.title('Mean Number of Votes by Rating Group')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Barplot rataâ€‘rata ROI per Country (â‰¥Â 50 film)
country_stats = (df.groupby('Country')['ROI']
                 .agg(['mean', 'count'])
                 .query('count >= 50')
                 .sort_values('mean', ascending=False)
                 .head(12))

plt.figure(figsize=(8, 5))
sns.barplot(x=country_stats['mean'], y=country_stats.index, orient='h', color='steelblue')
plt.title('Average ROI per Country (â‰¥50 films)')
plt.xlabel('Average ROI'); plt.ylabel('')
min_roi = country_stats['mean'].min()
max_roi = country_stats['mean'].max()
plt.xlim(min_roi - 0.2, max_roi + 0.2)
plt.tight_layout()
plt.show()

````

````python
# VOLUSI INDUSTRI PER TAHUN

yearly = (df.groupby('Year')
            .agg(film_count=('Title', 'count'),
                 mean_budget=('Budget_USD', 'mean'))
            .reset_index())

fig, ax1 = plt.subplots(figsize=(8, 4))
ax1.bar(yearly['Year'], yearly['film_count'], color='lightgray', alpha=.7)
ax1.set_xlabel('Year'); ax1.set_ylabel('# Films', color='gray')
ax2 = ax1.twinx()
ax2.plot(yearly['Year'], yearly['mean_budget'], linewidth=2)
ax2.set_ylabel('Mean Budget (USD)')

plt.title('Film Production & Mean Budget per Year')
plt.tight_layout(); plt.show()

# Boxplot Rating per Tahun
# Buat kolom baru range 5 tahunan
df['Year_Group'] = (df['Year'] // 5) * 5  # kelompok per 5 tahun
avg_rating_by_year = df.groupby("Year_Group")["Rating"].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=avg_rating_by_year, x="Year_Group", y="Rating", marker="o")
plt.title("Average Rating per 5-Year Group")
plt.ylim(6.8, 6.99)
plt.ylabel("Average Rating"); plt.xlabel("Year Group")
plt.grid(True)
plt.show()

# Line mean Awards per Tahun
mean_award = df.groupby('Year')['Num_Awards'].mean()
plt.figure(figsize=(8, 4))
mean_award.plot(marker='o')
plt.title('Average Number of Awards per Year')
plt.xlabel('Year'); plt.ylabel('Mean Awards')
plt.tight_layout(); plt.show()
````

````python
# SEGMENTASI PASAR
# Pieâ€¯Content Rating
df['Content_Rating'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribusi Content Rating")
plt.ylabel("")
plt.show()

# Bar AvgÂ BoxOffice per Country (â‰¥Â 50 film)
country_bo = (df.groupby('Country')['BoxOffice_USD']
              .agg(['mean', 'count'])
              .query('count >= 50')
              .sort_values('mean', ascending=False)
              .head(12))

country_bo['mean_million'] = country_bo['mean'] / 1_000_000

plt.figure(figsize=(8, 5))
sns.barplot(x=country_bo['mean_million'], y=country_bo.index, orient='h')
plt.title('Average Box Office per Country (â‰¥50 films)')
plt.xlabel('Average Box Office (Million USD)')
plt.xlim(500, 525)
plt.ylabel('')
plt.tight_layout()
plt.show()

# Bar count Film per Language (Topâ€‘15)
top_lang = df['Language'].value_counts().head(15)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_lang.values, y=top_lang.index, orient='h')
plt.title('Film Count per Language (Topâ€‘15)')
plt.xlim(7000, 7250)
plt.xlabel('# Films'); plt.ylabel('Language')
plt.tight_layout(); plt.show()
````

````python
from scipy.stats import kruskal

groups = [group["Rating"].values for name, group in df.groupby("Critic_Reviews")]
stat, p = kruskal(*groups)
print("Kruskal-Wallis Test: p =", p)
````

````python
review_stats = (df.groupby('Review_bin')['Rating']
                  .agg(['mean', 'count', 'std'])
                  .assign(se=lambda d: d['std'] / np.sqrt(d['count']),
                          ci_low=lambda d: d['mean'] - 1.96*d['se'],
                          ci_hi=lambda d: d['mean'] + 1.96*d['se'])
                  .reset_index())

plt.figure(figsize=(6, 4))
x = np.arange(len(review_stats))

plt.plot(x, review_stats['mean'], marker='o', color='teal', label='Mean Rating')

plt.fill_between(x, review_stats['ci_low'], review_stats['ci_hi'],
                 color='teal', alpha=0.2, label='95% CI')

plt.xticks(x, review_stats['Review_bin'])

plt.xlabel('Critic Reviews (binned)')
plt.ylabel('Mean Rating')
plt.title('Trend of Rating by Critic Reviews Group')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
df['Votes_decile'] = pd.qcut(df['Votes'], 10, labels=[f'D{i}' for i in range(1,11)])
import scipy.stats as st
pop_stats = (df.groupby('Votes_decile')['Rating']
               .agg(['mean', 'count', 'std'])
               .assign(se=lambda d: d['std']/np.sqrt(d['count']),
                       ci_low=lambda d: d['mean'] - 1.96*d['se'],
                       ci_hi=lambda d: d['mean'] + 1.96*d['se'])
               .reset_index())

plt.figure(figsize=(8, 4))
sns.lineplot(data=pop_stats, x='Votes_decile', y='mean', marker='o', color='steelblue', label='Mean Rating')
plt.fill_between(pop_stats['Votes_decile'],
                 pop_stats['ci_low'],
                 pop_stats['ci_hi'],
                 color='steelblue', alpha=0.2, label='95% CI')

plt.ylabel('Mean Rating')
plt.xlabel('Votes Decile (Q1 = sepi, Q10 = super-populer)')
plt.title('Trend: Mean Rating by Popularity Decile (Â±95% CI)')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
roi_stats = (df.groupby('Budget_Q')['ROI']
               .agg(['mean', 'count', 'std'])
               .assign(se=lambda d: d['std'] / np.sqrt(d['count']),
                       ci_low=lambda d: d['mean'] - 1.96*d['se'],
                       ci_hi=lambda d: d['mean'] + 1.96*d['se'])
               .reset_index())

plt.figure(figsize=(7, 4))
x = np.arange(len(roi_stats))

plt.plot(x, roi_stats['mean'], marker='o', color='indianred', label='Mean ROI')
plt.fill_between(x, roi_stats['ci_low'], roi_stats['ci_hi'],
                 color='indianred', alpha=0.25, label='95% CI')
plt.xticks(x, roi_stats['Budget_Q'])
plt.xlabel('Budget Quintile (Q1 = Murah â€¦ Q5 = Mahal)')
plt.ylabel('Mean ROI')
plt.title('Mean ROI by Budget Quintile')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
award_bins = pd.cut(df['Num_Awards'], bins=[-1,0,5,10,20], labels=['0','1â€‘5','6â€‘10','>10'])
award_stats = (df.groupby(award_bins)['ROI']
                 .agg(['mean', 'count', 'std'])
                 .assign(se=lambda d: d['std']/np.sqrt(d['count']),
                         ci_low=lambda d: d['mean'] - 1.96*d['se'],
                         ci_hi=lambda d: d['mean'] + 1.96*d['se'])
                 .reset_index())

plt.figure(figsize=(6, 4))
x = np.arange(len(award_stats))

plt.plot(x, award_stats['mean'], marker='o', color='darkgreen', label='Mean ROI')
plt.fill_between(x, award_stats['ci_low'], award_stats['ci_hi'],
                 color='darkgreen', alpha=0.25, label='95% CI')

plt.xticks(x, award_stats['Num_Awards'])
plt.xlabel('# Awards Category')
plt.ylabel('Mean ROI')
plt.title('Trend of ROI by Award Group')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
year_stat = (df.groupby('Year')['Rating']
               .agg(median='median',
                    q1=lambda s: s.quantile(.25),
                    q3=lambda s: s.quantile(.75))
               .reset_index())

plt.figure(figsize=(9, 4))

plt.fill_between(year_stat['Year'], year_stat['q1'], year_stat['q3'], alpha=0.2, label='IQR')
plt.plot(year_stat['Year'], year_stat['median'], marker='o', color='steelblue', label='Median')

plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Median Rating per Year (Shaded = IQR)')

plt.ylim(6.5, 7.2)

plt.legend()
plt.tight_layout()
plt.show()

````
